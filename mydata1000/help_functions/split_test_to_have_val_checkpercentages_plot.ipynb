Split the test data to validation and test

```python
from sklearn.model_selection import train_test_split
import pandas as pd

# Read the dataset into a DataFrame
df = pd.read_csv('mydata1000_test.csv', header=None)  

# Split the dataset into two sets based on the values in the first column
df_0 = df[df.iloc[:, 0] == 0]  
df_1 = df[df.iloc[:, 0] == 1]
df_2 = df[df.iloc[:, 0] == 2]
df_3 = df[df.iloc[:, 0] == 3]

# Split each set into training and testing sets with equal proportions
df_0_val, df_0_test = train_test_split(df_0, test_size=0.5, random_state=42)
df_1_val, df_1_test = train_test_split(df_1, test_size=0.5, random_state=42)
df_2_val, df_2_test = train_test_split(df_2, test_size=0.5, random_state=42)
df_3_val, df_3_test = train_test_split(df_3, test_size=0.5, random_state=42)

# Concatenate the training and testing sets
df_val = pd.concat([df_0_val, df_1_val, df_2_val, df_3_val], axis=0)
df_test = pd.concat([df_0_test, df_1_test, df_2_test, df_3_test], axis=0)

# Export the training and testing sets to CSV files
df_val.to_csv('val_dataset.csv', index=False, header=None)
df_test.to_csv('test_dataset.csv', index=False, header=None)

```

Check if percentages in val and test are similar

```python
import csv

# Define the file path
file_path = "val_dataset.csv"  


count_0 = 0
count_1 = 0
count_2 = 0
count_3 = 0


with open(file_path, 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        if row:  # Check if the row is not empty
            # Get the first column value from the row
            value = row[0]
            # Iterate over the characters in the value and count occurrences of each digit
            for digit in value:
                if digit == '0':
                    count_0 += 1
                elif digit == '1':
                    count_1 += 1
                elif digit == '2':
                    count_2 += 1
                elif digit == '3':
                    count_3 += 1

# Output the counts
print("Occurrences of each digit:")
print("0:", count_0/150*100)
print("1:", count_1/150*100)
print("2:", count_2/150*100)
print("3:", count_3/150*100)
```
